```python
# Reading data
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv')

data.head(2)

df = data.copy()
df.dropna() # Elimina todos los registros vacios

df2 = data.copy()
df2.dropna(subset=['X1']) # Elimina todos los registros vacios de una sola columna o una lista

```

```python
# Convert columns

data['X1_transform'] = data['X1'] / 100
data.head()

#data.dtypes # Obtener los tipos de datos por columna
data['X1_transform'] = data['X1_transform'].astype('int')
data.dtypes # Obtener los tipos de datos por columna
```

```python
# Normalizando datos en pandas
df3 = data.copy()
df3['X1'] = (df3['X1'] - df3['X1'].min()) / (df3['X1'].max() - df3['X1'].min())
df3.head()
```
