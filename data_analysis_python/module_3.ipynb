### Estadística descriptiva

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
df = pd.read_csv('data/new_df.csv')
df = df.dropna()
df['Y'] = df['Y'].apply(lambda x: int(x))
df.head(2)
```

El metodo describe de pandas nos permite obtener información estadística del dataframe

```python
df.describe()
```

Obtener la suma de las categorías de alguna variable, en este caso la columna Y

```python
df['Y'].value_counts()
```

El scatterplot muestra la relación entre dos variables

```python
plt.scatter(df['X11'], df['X10'])
plt.title('Relacion entre X11 y X10')
plt.xlabel('X11')
plt.ylabel('X10')
```

La función groupby de pandas nos permite conocer si hay alguna relación con los datos de la categoría, en este caso vamos a ver si hay alguna relación con la variable Y y X11 y podemos observar el promedio de X11 de acuerdo a cada categoría dentro de la variable Y

```python
df2 = df[['Y', 'X11']]
df2.groupby(['Y'], as_index=False).mean()
```

#### Analisis de Varianza ANOVA

Permite conocer la variación estadística entre grupos

El análisis de varianza se utiliza para determinar si existe correlación entre diferentes grupos de una variable categórica. Como resultado obtenemos un valor f-test y el valor p

Un f-test pequeño implica una correlación pobre entre las categorías y la variable objetivo
Un f-test alto implica una correlación fuerte entre las categorías y la variable objetivo

```python
df_anova = df[['make', 'price']]
grouped_anova = df_anova.groupby(['make'])
anova_results = stats.f_oneway(grouped_anova.get_group('honda')['price'], grouped_anova.get_group('subaru')['price'])
```

#### Correlación

La correlación de pearson es una medida que nos permite determinar la fuerza de correlación entre dos variables.
* Valor cernano a uno nos indica que la correlación es positiva, a medida que aumenta una la otra igual aumenta
* Valor cernano a cero indica que no hay correlación lineal
* Valor cercano a menos uno indica que la correlación es negativa, a medida que aumenta una la otra disminuye

```python
plt.figure(figsize=[20,10])
sns.heatmap(df.corr(), annot=True)
```
